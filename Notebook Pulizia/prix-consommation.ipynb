# **1.Data Preparation**

```python
path = '/content/drive/Shareddrives/Progetto Big Data/DataSet/economicref-france-indice-prix-consommation.csv'
df = pd.read_csv(path, sep = ';')
```

## 1.1.Gestione dei tipi

```python
df.dtypes
```

```python
df[["Code", "Type",'Sous-type', 'Base', 'Population' ,
    'Zone concernée', 'Code COICOP', 'Désignation COICOP' ]] = df[["Code", "Type",'Sous-type', 'Base', 'Population' ,
                                                                   'Zone concernée', 'Code COICOP', 'Désignation COICOP' ]].astype("string")

df['Période'] = pd.to_datetime(df['Période'])
```

```python
df.dtypes
```

## 1.2.Comprensione del dataSet e sistemazione delle colonne

```python
df = df.drop('Indice',axis = 1)
df.head()
```

```python
df.describe(include = 'all')
```

```python
df.info()
```

```python
df = df[(df['Période'] >= '2022-04-01')]
```

Eslcudo i valori provvisori con codice P, considero solo i valori normali A

```python
#Escludo i valori provvisori
df = df[(df['Code'] == 'A')]
df
```
