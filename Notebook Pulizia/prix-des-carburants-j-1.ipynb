# **1.Data Preparation**

```python
path = '/content/drive/Shareddrives/Progetto Big Data/DataSet/prix-des-carburants-j-1.csv'
df_esteso = pd.read_csv(path, sep = ';')
```

```python
df_esteso.head()
```

## 1.1 Eliminazione dei dipartimenti del gruppo 1

```python
df = df_esteso.drop(df_esteso[df_esteso['Nom Officiel Région'] == 'Hauts-de-France'].index)
df = df.drop(df[df['Nom Officiel Région'] == 'Grand Est'].index)
df = df.drop(df[df['Nom Officiel Région'] == 'Île-de-France'].index)
df = df.drop(df[df['Nom Officiel Région'] == 'Normandie'].index)
df = df.drop(df[df['Nom Officiel Région'] == 'Bretagne'].index)
df = df.drop(df[df['Nom Officiel Région'] == 'Bourgogne-Franche-Comté'].index)
df = df.drop(df[df['Nom Officiel Région'] == 'Centre-Val de Loire'].index)
```

## 1.2.Gestione dei valori nulli

```python
#Rimuovo le righe con il nome della regione nullo
df = df.dropna(subset=['Nom Officiel Région'])
```

```python
missing_data = df.isnull()
for column in missing_data.columns.values.tolist():
    print (missing_data[column].value_counts())
    print("")
```

## 1.3.Gestione dei tipi



```python
df = df.rename(columns={'Nom Officiel Commune / Arrondissement Municipal Majuscule': 'Nom Commune'})
```

```python
df[['Adresse','Nom Commune','Carburant','Code Officiel Département',
    'Nom Officiel Département','Code Officiel Région','Nom Officiel Région']] =df[['Adresse','Nom Commune','Carburant','Code Officiel Département',
                                                                                   'Nom Officiel Département','Code Officiel Région','Nom Officiel Région']].astype("string")
df['Mise à jour'] = pd.to_datetime(df['Mise à jour'], utc=True)
```

## 1.4.Comprensione del dataSet e sistemazione delle colonne

```python
#Elimino le righe che non mi servono
colonne_da_elim = ['Code Postal','Présence','Automate 24-24','Timetable','Rupture','Services',
                   'Marque','Nom','Geo Point','Code officiel commune ou arrondissement','Code Officiel EPCI','Nom Officiel EPCI']
df = df.drop(colonne_da_elim, axis=1)
```

```python
#Riporto tutti i prezzi in euro/litro
df_prezzi = df[['Prix Gazole','Prix SP95','Prix SP98','Prix GPLc','Prix E10','Prix E85']]
maschera = df_prezzi < 0.01
df_prezzi = df_prezzi.mask(maschera, df_prezzi * 1000)
df[['Prix Gazole','Prix SP95','Prix SP98','Prix GPLc','Prix E10','Prix E85']] = df_prezzi
df
```

```python
lista_nome_regioni = df['Nom Officiel Région'].unique()
print(lista_nome_regioni)
print(df.shape)
```

```python
date_aprile_2022 = pd.Timestamp('2022-04-01', tz='UTC')
is_before = df['Mise à jour'] < date_aprile_2022
df =  df.drop(df[is_before].index)
```

```python
df.info()
```

```python
df_raggruppato = df.groupby(df['Mise à jour'].dt.date)[['Prix Gazole','Prix SP95','Prix SP98','Prix GPLc','Prix E10','Prix E85']].mean().reset_index()
df_raggruppato
```
