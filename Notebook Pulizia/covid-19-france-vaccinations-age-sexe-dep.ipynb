# **1.Data Preparation**

```python
path = '/content/drive/Shareddrives/Progetto Big Data/DataSet/covid-19-france-vaccinations-age-sexe-dep.csv'
df = pd.read_csv(path, sep = ';')
```

## 1.1.Comprensione del dataSet e sistemazione delle colonne

```python
df.head(500)
```

```python
#Prendo solo i dati relativi a tutte le età
df = df[df["Libellé de la variable"] == "Tous âges"]
```

```python
#Prendo i dati più recenti del dipartimento
df= df.sort_values(by=['Date']).drop_duplicates(subset='Nom Officiel Département', keep='last')
```

```python
#Sommo tutti i vaccinati con almeno una dose
column_names = ['Nombre cumulé de doses n°1', 'Nombre cumulé de vaccinations complètes (doses n°2)', 'Nombre cumulé de personnes vaccinées avec une dose de rappel (3e dose)', 'Nombre cumulé de personnes vaccinées avec une deuxième dose de rappel (4e dose)','Nombre cumulé de personnes vaccinées avec une dose de rappel bivalent','Nombre cumulé de personnes ayant reçu une troisième dose de rappel']
df['Nombre total de vaccinations']= df[column_names].sum(axis=1)
```

```python
#Faccio la media delle varie percentuali di vaccinazioni
column_names_perc = ['Couverture vaccinale partielle (dose n°1)', 'Couverture vaccinale complète (dose n°2)']
df['Couverture moyenne 1/2']= df[column_names_perc].mean(axis=1)
```

```python
#Rimuovo colonne inutili
columns = ['Type de variable','Libellé de la variable']
df = df.drop(columns, axis=1)
```

## 1.2.Gestione dei valori nulli

```python
missing_data = df.isnull()
missing_data.head()
```

## 1.3.Gestione dei tipi

```python
df.dtypes
```

```python
df[['Nom Officiel Région','Code Iso 3166-3 Zone','Nom Officiel Département','Code Officiel Département']] = df[['Nom Officiel Région','Code Iso 3166-3 Zone','Nom Officiel Département','Code Officiel Département']].astype("string")
df['Date'] = pd.to_datetime(df['Date'])
```

## 1.4 Eliminazione dei dipartimenti del gruppo 1

```python
df.shape
```

```python
df = df.drop(df[df['Nom Officiel Région'] == 'Hauts-de-France'].index)
df = df.drop(df[df['Nom Officiel Région'] == 'Grand Est'].index)
df = df.drop(df[df['Nom Officiel Région'] == 'Île-de-France'].index)
df = df.drop(df[df['Nom Officiel Région'] == 'Normandie'].index)
df = df.drop(df[df['Nom Officiel Région'] == 'Bretagne'].index)
df = df.drop(df[df['Nom Officiel Région'] == 'Bourgogne-Franche-Comté'].index)
df = df.drop(df[df['Nom Officiel Région'] == 'Centre-Val de Loire'].index)
```

```python
lista_dipartimenti = df['Nom Officiel Département'].unique()
print(lista_dipartimenti)
lista_dipartimenti.size
```
