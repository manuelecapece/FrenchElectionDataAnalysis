# **1.Data Preparation**

Read del dataSet da Drive

```python
path = '/content/drive/Shareddrives/Progetto Big Data/DataSet/elections-france-presidentielles-2022-1er-tour-par-bureau-de-vote.csv'
df = pd.read_csv(path, sep = ';')
```

## 1.1.Comprensione del dataSet e sistemazione delle colonne

```python
df.head(500)
```

```python
df['% Voix/Exp'] = df['% Voix/Exp'].astype('string')
df['% Voix/Exp'] = df['% Voix/Exp'].replace(',', '.',regex = True)
df.head()
```

```python
df.describe(include = 'all')
```

```python
df.info()
```

```python
df_separato = df['location'].str.split(',', expand=True)
df_separato.columns = ['Latitudine', 'Longitudine']
df_separato.head()
```

```python
df = pd.concat([df, df_separato], axis=1)
df = df.drop('location',axis = 1)
df.head()
```

```python
duplicati_presenti = df.duplicated().any()

if duplicati_presenti:
    print("Sono presenti duplicati nel DataFrame.")
else:
    print("Non ci sono duplicati nel DataFrame.")
```

## 1.2.Gestione dei valori nulli

```python
#Rimuovo le righe con il nome della regione nullo
df = df.dropna(subset=['Nom Officiel Région'])
```

```python
missing_data = df.isnull()
missing_data.head()
```

```python
for column in missing_data.columns.values.tolist():
    print (missing_data[column].value_counts())
    print("")
```

```python
df = df.drop('lib_du_b_vote',axis = 1)
```

## 1.3.Gestione dei tipi

```python
df.dtypes
```

```python
df[['Code du département','Code de la commune', 'Code du b.vote']] = df[['Code du département','Code de la commune', 'Code du b.vote']].astype("string")
df[['Libellé du département','Libellé de la circonscription', 'Libellé de la commune',
    'Sexe','Nom','Prénom','Nom Officiel Région','scrutin_code']] = df[['Libellé du département','Libellé de la circonscription', 'Libellé de la commune',
                                                        'Sexe','Nom','Prénom','Nom Officiel Région','scrutin_code']].astype("string")
df[['Latitudine','Longitudine','% Voix/Exp']] = df[['Latitudine','Longitudine','% Voix/Exp']].astype('float')
```

```python
df.dtypes
```

## 1.4 Eliminazione dei dipartimenti del gruppo 1

```python
df.shape
```

```python
lista_dipartimenti = df['Libellé du département'].unique()
print(lista_dipartimenti)
lista_dipartimenti.size
```

```python
df = df.drop(df[df['Nom Officiel Région'] == 'Hauts-de-France'].index)
df = df.drop(df[df['Nom Officiel Région'] == 'Grand Est'].index)
df = df.drop(df[df['Nom Officiel Région'] == 'Île-de-France'].index)
df = df.drop(df[df['Nom Officiel Région'] == 'Normandie'].index)
df = df.drop(df[df['Nom Officiel Région'] == 'Bretagne'].index)
df = df.drop(df[df['Nom Officiel Région'] == 'Bourgogne-Franche-Comté'].index)
df = df.drop(df[df['Nom Officiel Région'] == 'Centre-Val de Loire'].index)
```

```python
lista_nome_regioni = df['Nom Officiel Région'].unique()
print(lista_nome_regioni)
print(df.shape)
```

```python
df = df.sort_values('Libellé du département')
lista_dipartimenti = df['Libellé du département'].unique()
print(lista_dipartimenti)
lista_dipartimenti.size
```

## 1.7.Crezione del file excel dataSet inflazionePerDipartimento

```python
df_inflazione = df[['Code du département','Libellé du département','Nom Officiel Région']].sort_values('Libellé du département')
df_inflazione = df_inflazione.drop_duplicates()
#Drop delle isole con dati relativi all'inflazione non disponibili
df_inflazione = df_inflazione.drop(df_inflazione[df_inflazione['Code du département'] == '97'].index)
#Drop dipartimenti duplicati
df_inflazione = df_inflazione.drop(df_inflazione[df_inflazione['Code du département'] == '01'].index)
df_inflazione = df_inflazione.drop(df_inflazione[df_inflazione['Code du département'] == '03'].index)
df_inflazione = df_inflazione.drop(df_inflazione[df_inflazione['Code du département'] == '04'].index)
df_inflazione = df_inflazione.drop(df_inflazione[df_inflazione['Code du département'] == '05'].index)
df_inflazione = df_inflazione.drop(df_inflazione[df_inflazione['Code du département'] == '06'].index)
df_inflazione = df_inflazione.drop(df_inflazione[df_inflazione['Code du département'] == '07'].index)
df_inflazione = df_inflazione.drop(df_inflazione[df_inflazione['Code du département'] == '09'].index)
df_inflazione['Indice des prix dans le département'] = np.nan
```

```python
lista_dipartimenti = df_inflazione['Libellé du département'].unique()
print(lista_dipartimenti)
lista_dipartimenti.size
```

```python
path = '/content/drive/Shareddrives/Progetto Big Data/DataSet Puliti/inflazionePerDipartimento.xlsx'
df_inflazione.to_excel(path , index=False);
```
